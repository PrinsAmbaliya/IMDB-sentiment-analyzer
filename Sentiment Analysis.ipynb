```python
import pandas as pd 
```

```python
dataset = pd.read_csv("IMDB Dataset.csv")
dataset.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
dataset["sentiment"] =  dataset["sentiment"].replace({"positive": 1, "negative": 0}).values
```

```python
dataset.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer()
```

```python
X_counts = vect.fit_transform(dataset['review'].astype(str))
```

```python
docs = np.array(['i am good persone'
               'you are my best friend'
               'hello chhagan how are you'])
```

```python
bag = vect.fit_transform(docs)
```

```python
print(vect.vocabulary_)
```

```python
print(bag.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
np.set_printoptions(precision=2)
tfid=TfidfTransformer(use_idf=True,norm='l2',smooth_idf=True)
print(tfid.fit_transform(bag).toarray())
```

```python
import nltk
nltk.download('stopwords')
```

```python
y = dataset['sentiment'].values
x = tfid.fit_transform(X_counts)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.5)
```

```python
import pickle
from sklearn.linear_model import LogisticRegressionCV
```

```python
clf = LogisticRegressionCV(
    cv=5,
    scoring='accuracy',
    random_state=0,
    n_jobs=-1,
    verbose=1,
    max_iter=1000,
    solver='saga'
).fit(x_train, y_train)

with open('save_model.sav', 'wb') as save_model:
    pickle.dump(clf, save_model)
```

```python
filename = 'save_model.sav'
save_clf = pickle.load(open(filename,'rb'))
save_clf.score(x_test,y_test)*100
```
